# ggplot correlation  with values shown

```r
df = read.csv("Cas9_affinity_function_HS2_7gRNAs_reformat.txt",sep="\t")
```

```r
head(df)
```

```r
library(ggpubr)
```

```r
library(ggplot2)
ggplot(df,aes(x = Cas9_affinity, y = HBE)) +
geom_smooth(data = subset(df, cat %in% c("HBE")),fill="blue",method='lm')+
geom_point(data = subset(df, cat %in% c("HBE")),color="blue")+
geom_smooth(data = subset(df, cat %in% c("HBG")),fill="green",method='lm')+
geom_point(data = subset(df, cat %in% c("HBG")),color="green")+
geom_smooth(data = subset(df, cat %in% c("HBB")),fill="red",method='lm')+
geom_point(data = subset(df, cat %in% c("HBB")),color="red")
# +
# stat_cor(data = subset(df, cat %in% c("HBE")),method = "pearson", label.x = 3, label.y = 30,color="red")
```

```r
library(ggplot2)
ggplot(df,aes(x = Cas9_affinity, y = HBE)) +
geom_smooth(data = subset(df, cat %in% c("HBE")),fill="blue",color="blue",method='lm')+
geom_point(data = subset(df, cat %in% c("HBE")),color="blue")+
stat_cor(data = subset(df, cat %in% c("HBE")),method = "pearson", label.x = 250, label.y = 0.95,color="blue")+
geom_smooth(data = subset(df, cat %in% c("HBG")),fill="green",color="green",method='lm')+
geom_point(data = subset(df, cat %in% c("HBG")),color="green")+
stat_cor(data = subset(df, cat %in% c("HBG")),method = "pearson", label.x = 250, label.y = 0.85,color="green")+
geom_smooth(data = subset(df, cat %in% c("HBB")),fill="red",color="red",method='lm')+
geom_point(data = subset(df, cat %in% c("HBB")),color="red")+
stat_cor(data = subset(df, cat %in% c("HBB")),method = "pearson", label.x = 250, label.y = 0.75,color="red")
```

```r
library(ggplot2)
library(ggpubr)
ggplot(df,aes(x = Cas9_affinity, y = HBE)) +
geom_smooth(data = subset(df, cat %in% c("HBE")),fill="blue",se=FALSE,color="blue",method='lm')+
geom_point(data = subset(df, cat %in% c("HBE")),color="blue")+
stat_cor(data = subset(df, cat %in% c("HBE")),method = "pearson", label.x = 250, label.y = 0.95,color="blue")+
geom_smooth(data = subset(df, cat %in% c("HBG")),fill="green",se=FALSE,color="green",method='lm')+
geom_point(data = subset(df, cat %in% c("HBG")),color="green")+
stat_cor(data = subset(df, cat %in% c("HBG")),method = "pearson", label.x = 250, label.y = 0.85,color="green")+
geom_smooth(data = subset(df, cat %in% c("HBB")),fill="red",se=FALSE,color="red",method='lm')+
geom_point(data = subset(df, cat %in% c("HBB")),color="red")+
stat_cor(data = subset(df, cat %in% c("HBB")),method = "pearson", label.x = 250, label.y = 0.75,color="red")+
theme_classic()+
ggsave("figure.png",dpi=300)
```

```r
library(ggplot2)
library(ggpubr)
ggplot(df,aes(x = Cas9_affinity, y = HBE)) +
geom_smooth(data = subset(df, cat %in% c("HBE")),fill="blue",se=FALSE,color="blue",method='lm')+
geom_point(data = subset(df, cat %in% c("HBE")),color="blue")+
stat_cor(data = subset(df, cat %in% c("HBE")),method = "pearson", label.x = 250, label.y = 0.95,color="blue",aes(label = paste('HBE',..r.label.., ..p.label.., sep = "~`,`~")))+
geom_smooth(data = subset(df, cat %in% c("HBG")),fill="green",se=FALSE,color="green",method='lm')+
geom_point(data = subset(df, cat %in% c("HBG")),color="green")+
stat_cor(data = subset(df, cat %in% c("HBG")),method = "pearson", label.x = 250, label.y = 0.85,color="green",aes(label = paste('HBG',..r.label.., ..p.label.., sep = "~`,`~")))+
geom_smooth(data = subset(df, cat %in% c("HBB")),fill="red",se=FALSE,color="red",method='lm')+
geom_point(data = subset(df, cat %in% c("HBB")),color="red")+
stat_cor(data = subset(df, cat %in% c("HBB")),method = "pearson", label.x = 250, label.y = 0.75,color="red",aes(label = paste('HBB',..r.label.., ..p.label.., sep = "~`,`~")))+
theme_classic()+
ggsave("figure.png",dpi=300)
```
